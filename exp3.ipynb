```python
import numpy as np

# Define matrices
x = np.array([[1, 2], [3, 4]], dtype=np.float64)
y = np.array([[6, 9], [4, 4]], dtype=np.float64)

# Print matrices
print("x =", x)
print("y =", y)

# Element-wise operations
print("Element-wise addition:", np.add(x, y))
print("Element-wise subtraction:", np.subtract(x, y))
print("Element-wise multiplication:", np.multiply(x, y))
print("Element-wise square root of x:", np.sqrt(x))

# Matrix multiplication
print("Matrix multiplication:", np.dot(x, y))

# Sums
print("Sum of all elements of matrix x:", np.sum(x))
print("Sum of elements in each column of matrix y:", np.sum(y, axis=0))
print("Sum of elements in each row of matrix y:", np.sum(y, axis=1))

# Transpose of matrix x
print("Transpose of matrix x:", x.T)
```

```python
import numpy as np

# Example for broadcasting
print("Example for broadcasting: ")

# Define vectors
v = np.array([1, 2, 3])
w = np.array([4, 5])

print("v =", v)
print("w =", w)

# Outer product using broadcasting
# Reshape v to (3, 1) and broadcast with w to get the outer product
print("Outer product of above vectors:")
print(np.reshape(v, (3, 1)) * w)  # Reshape v and multiply by w

# Define matrix
x = np.array([[1, 2, 3], [4, 5, 6]])

print("x =", x)

# Adding vector v to each row of matrix x
print("Resultant matrix after adding the vector v to each row of matrix x:")
print(x + v)  # Broadcasting v to each row of x

# Example for broadcasting fails
print("Example for broadcasting fails:")

print("Adding the vector w to each column matrix x will generate an error")

# Solution: Reshape w to be compatible for broadcasting
print("Solution: Reshape the matrix w then the result will be:")
print(x + np.reshape(w, (2, 1)))  # Reshape w to (2, 1) to match dimensions for broadcasting
```

```python
import numpy as np
from scipy.linalg import svd

# Define matrix A
A = np.array([[1, 2], [3, 4], [5, 6]])

# Print matrix A and its shape
print("A =\n", A)
print("Shape of matrix A:", A.shape)
print("")

# Perform Singular Value Decomposition
U, s, VT = svd(A)

# Print results of SVD
print("U =\n", U)
print("Shape of matrix U:", U.shape)
print("")

print("Sigma (diagonal matrix), s =\n", s)
print("Shape of matrix s:", s.shape)
print("")

print("VT =\n", VT)
print("Shape of matrix VT:", VT.shape)
print("")

# Construct Sigma as a diagonal matrix
sigma = np.zeros((A.shape[0], A.shape[1]))
sigma[:len(s), :len(s)] = np.diag(s)

# Reconstruct the matrix B from U, Sigma, and VT
B = np.dot(U, np.dot(sigma, VT))

# Print the reconstructed matrix
print("Reconstructed matrix:\n", B)
```

```python
import numpy as np
from scipy.linalg import pinv
from numpy.linalg import svd

# Define matrix A
A = np.array([[1, 2], [3, 4], [5, 6]])

# Print matrix A
print("A =\n", A)

# Compute pseudoinverse using scipy's pinv function
print("Pseudoinverse of matrix A calculated by function pinv is:\n", pinv(A))

# Perform Singular Value Decomposition
U, s, VT = svd(A)

# Compute pseudoinverse manually
# Create the diagonal matrix D with reciprocal of non-zero singular values
d = 1.0 / s
D = np.zeros(A.shape[::-1])
D[:len(s), :len(s)] = np.diag(d)

# Compute the pseudoinverse B using U, D, and VT
B = VT.T.dot(D.T).dot(U.T)

# Print the pseudoinverse computed using SVD
print("Pseudoinverse of matrix A calculated by using svd is:\n", B)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot sine and cosine
plt.plot(x, y_sin, label='Sine')
plt.plot(x, y_cos, label='Cosine')

# Add labels and title
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')

# Add legend
plt.legend()

# Show the plot
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create subplots
plt.figure(figsize=(8, 6))

# First subplot: Sine wave
plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st subplot
plt.plot(x, y_sin, 'r')  # 'r' is the color red
plt.title('Sine')

# Second subplot: Cosine wave
plt.subplot(2, 1, 2)  # 2 rows, 1 column, 2nd subplot
plt.plot(x, y_cos, 'b')  # 'b' is the color blue
plt.title('Cosine')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()

```
